```python
import nltk
from nltk.corpus import gutenberg, brown
from urllib import request
from bs4 import BeautifulSoup
from nltk import word_tokenize, Text, FreqDist
import matplotlib.pyplot as plt
from wordcloud import WordCloud

nltk.download('all')


print("Gutenberg Texts:", gutenberg.fileids())
```

```python
def analyze_gutenberg():
    for fileid in gutenberg.fileids():
        raw_text = gutenberg.raw(fileid)
        words = gutenberg.words(fileid)
        sentence = gutenberg.sents(fileid)
        total_words = len(words)
        total_chars = len(raw_text)
        total_sentence = len(sentence)
        avg_chars_per_word = total_chars / total_words
        ttr = len(set(words)) / total_words
        print(f"\nText: {fileid}")
        print("Total Words:", total_words)
        print("Total Characters:", total_chars)
        print("Total Sentences:", total_sentence)
        print("Average Characters per Word:", avg_chars_per_word)
        print("Type Token Ratio (Lexical Richness):", ttr)

analyze_gutenberg()
print("\nBrown Corpus Categories:", brown.categories())
```

```python

def analyze_brown():
    for category in brown.categories():
        words = brown.words(categories=category)
        sents = brown.sents(categories=category)
        print(f"\nCategory: {category}")
        print("Total Words:", len(words))
        print("Total Sentences:", len(sents))
        fdist = FreqDist(words)
        print("Most Common Words:", fdist.most_common(5))

analyze_brown()
```

```python
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')
tokens = word_tokenize(raw)

print("\nWeb Text Sample:", raw[1000:1200])
print("Total Tokens:", len(tokens))

text_obj = Text(tokens)
print("Collocations:")
text_obj.collocations()
print("\nConcordance for 'freedom':")
text_obj.concordance("freedom", lines=5)
```

```python
html_url = "http://www.gutenberg.org/files/1342/1342-h/1342-h.htm"
response = request.urlopen(html_url)
html = response.read().decode('utf8')
soup = BeautifulSoup(html, "html.parser")
html_text = soup.get_text()
tokens_html = word_tokenize(html_text)
text_html = Text(tokens_html)

print("\nHTML Token Sample:", tokens_html[:30])
print("\nConcordance for 'marriage':")
text_html.concordance("marriage", lines=5)
```

```python
local_text = """Natural Language Processing with NLTK is powerful and fun.
This is a sample document for testing tokenization and word cloud generation."""
tokens_local = word_tokenize(local_text)

print("\nLocal File Tokens:", tokens_local)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(local_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
