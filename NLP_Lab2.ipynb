```python
def Sent_to_list(sentence):
  sentence = sentence.lower()
  words = nltk.word_tokenize(sentence)
  words = [word for word in words if word not in stopwords]
  words = [re.sub('[^a-zA-Z0-9]+', '', word) for word in words]
  words = [word for word in words if word not in [""]]
  return (words)

words = Sent_to_list("India, country that occupies the greater part of South Asia. ")
vocab = sorted(set(words))
#print(vocab)
```

```python
onehot = []

for word in vocab:
  temp = [0 for _ in vocab]
  index = vocab.index(word)
  temp[index] = 1
  onehot.append(temp)

Final = {vocab[i]: onehot[i] for i in range(len(vocab))}
Final
```

```python
def countWords(doc):
  doc = Sent_to_list(doc)
  c = {}
  for word in doc:
    c[word] = doc.count(word)
  return (c)

def MissingWords(doc, vocab):
  for key in vocab:
    if key not in doc.keys():
      doc[key] = 0
  return doc
```

```python
#Bag of Words

doc1 = "Game of Thrones is an amazing tv series!"
doc2 = "Game of Thrones is the best tv series!"
doc3 = "Game of Thrones is so great"

vocab = set(Sent_to_list(doc1) + Sent_to_list(doc2) + Sent_to_list(doc3))


count_doc1 = MissingWords(countWords(doc1), vocab)
count_doc2 = MissingWords(countWords(doc2),vocab)
count_doc3 = MissingWords(countWords(doc3),vocab)
print(count_doc1)
pd.DataFrame([count_doc1,count_doc2,count_doc3])
```

```python
#TF

def get_tf(doc):
  tf = {}
  for key in doc.keys():
    tf[key] = doc[key]/ len(doc)
  return tf

tf_doc1 = get_tf(count_doc1)
tf_doc2 = get_tf(count_doc2)
tf_doc3 = get_tf(count_doc3)


pd.DataFrame([tf_doc1,tf_doc2,tf_doc3])

```

```python
#IDF

import math

def get_idf(all_docs):
    N = len(all_docs)
    idf_dict = {}
    for word in all_docs[0].keys():
        containing_docs = sum([1 for doc in all_docs if doc[word] > 0])
        idf_dict[word] = math.log10(N / (containing_docs))
    return idf_dict

idf = get_idf([count_doc1, count_doc2, count_doc3])
pd.DataFrame([idf])

```

```python
#TF-IDF

def get_tfidf(tf, idf):
    tfidf = {}
    for word in tf.keys():
        tfidf[word] = tf[word] * idf[word]
    return tfidf

tfidf_doc1 = get_tfidf(tf_doc1, idf)
tfidf_doc2 = get_tfidf(tf_doc2, idf)
tfidf_doc3 = get_tfidf(tf_doc3, idf)

pd.DataFrame([tfidf_doc1, tfidf_doc2, tfidf_doc3])
```
